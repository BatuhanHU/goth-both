```python
import matplotlib.pyplot as plt
import ast
import pandas as pd
import nltk
from nltk.corpus import stopwords
from collections import defaultdict
from nltk.tokenize import word_tokenize 
from tqdm import tqdm
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
tagDict = defaultdict(int)

```

```python
# Using readline()
file1 = open('booksummaries.txt', 'r',  encoding="utf8")

while True:
    # Get next line from file
    line = file1.readline()
    # if line is empty
    # end of file is reached
    if not line:
        break
    data = line.strip().split("\t")
    try:
        res = ast.literal_eval(data[5])
        for key in res.keys():
            tagDict[res[key]] += 1
    except:
        pass
file1.close()
```

```python
sortedDict = dict(reversed(sorted(tagDict.items(), key=lambda item: item[1], )))


fig = plt.figure(figsize=(15, 10))
ax = fig.add_axes([0,0,1,1])
tags = list(sortedDict.keys())[:10]
values = list(sortedDict.values())[:10]
ax.bar(tags,values)
plt.show()
```

```python

df = pd.read_csv('booksummaries.txt', delimiter="\t", names = ["WID", "Book Name","Author","Date","Tags","Summary" ], encoding="UTF-8")

emptyDf = df[df['Tags'].isna()]
validDf = df.dropna(subset=['Tags'], how='all')
#df["d"]

indx = 0
for i in validDf["Tags"]:
    i = ast.literal_eval(i)
    validDf.iat[indx,4] = i
    indx += 1
```

```python
train = validDf.sample(frac=0.8, random_state=42)  
# random state is a seed value
test = validDf.drop(train.index)
```

```python
indx = 0
stop_words = set(stopwords.words('english')) 

for i in tqdm(train["Summary"]):
    word_tokens = word_tokenize(i.lower()) 
    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
    train.iat[indx,5] = filtered_sentence
    indx += 1
```
